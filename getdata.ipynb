```python
#import the dependencies
import requests
import json
import pandas as pd
import re
```

```python
url = "https://data.cdc.gov/resource/dxpw-cm5u.json?$query=SELECT%0A%20%20%60stateabbr%60%2C%0A%20%20%60placename%60%2C%0A%20%20%60placefips%60%2C%0A%20%20%60population2010%60%2C%0A%20%20%60arthritis_crudeprev%60%2C%0A%20%20%60arthritis_crude95ci%60%2C%0A%20%20%60bphigh_crudeprev%60%2C%0A%20%20%60bphigh_crude95ci%60%2C%0A%20%20%60cancer_crudeprev%60%2C%0A%20%20%60cancer_crude95ci%60%2C%0A%20%20%60casthma_crudeprev%60%2C%0A%20%20%60casthma_crude95ci%60%2C%0A%20%20%60csmoking_crude95ci%60%2C%0A%20%20%60geolocation%60%2C%0A%20%20%60csmoking_crudeprev%60%2C%0A%20%20%60diabetes_crudeprev%60%2C%0A%20%20%60diabetes_crude95ci%60%2C%0A%20%20%60obesity_crudeprev%60%2C%0A%20%20%60obesity_crude95ci%60%0AWHERE%20caseless_one_of(%60stateabbr%60%2C%20%22CA%22%2C%20%22TX%22%2C%20%22NY%22)";
response =requests.get(url)
print(response)
```

```python
response_json= response.json()
print(json.dumps(response_json, indent=4, sort_keys=True))
```

```python
data_pd=pd.DataFrame(response_json)
data_pd.head()
```

```python
new_data_pd= [col for col in data_pd.columns if "prev" in col]
print(list(data_pd.columns))
print(new_data_pd)
```

```python
new_data_pd_2= data_pd.filter(regex="prev")
print(new_data_pd_2)
```

```python
new_data_pd_3= data_pd[["stateabbr","placename","placefips","population2010",'arthritis_crudeprev', 'bphigh_crudeprev', 'cancer_crudeprev', 'casthma_crudeprev', 'csmoking_crudeprev', 'diabetes_crudeprev', 'obesity_crudeprev']]
new_data_pd_3
```

```python
new_data_pd_3.dtypes
```

```python
new_data_pd_3["population2010"]=new_data_pd_3["population2010"].astype("double")
new_data_pd_3["arthritis_crudeprev"]=new_data_pd_3["arthritis_crudeprev"].astype("double")
new_data_pd_3["bphigh_crudeprev"]=new_data_pd_3["bphigh_crudeprev"].astype("double")
new_data_pd_3["cancer_crudeprev"]=new_data_pd_3["cancer_crudeprev"].astype("double")
new_data_pd_3["casthma_crudeprev"]=new_data_pd_3["casthma_crudeprev"].astype("double")
new_data_pd_3["csmoking_crudeprev"]=new_data_pd_3["csmoking_crudeprev"].astype("double")
new_data_pd_3["diabetes_crudeprev"]=new_data_pd_3["diabetes_crudeprev"].astype("double")
new_data_pd_3["obesity_crudeprev"]=new_data_pd_3["obesity_crudeprev"].astype("double")
new_data_pd_3.dtypes
```

```python
new_data_pd_3=new_data_pd_3.rename(columns={
    "arthritis_crudeprev": "Arthritis",
    "bphigh_crudeprev":"Blood Pressure High",
    "cancer_crudeprev":"Cancer",
    "casthma_crudeprev":"Asthma",
    "csmoking_crudeprev":"Smoking",
    "diabetes_crudeprev":"Diabetes",
    "obesity_crudeprev":"Obesity",
    "stateabbr":"state",
    "placename":"City",
    "population2010":"Population"
})
```

```python
final_data_1=pd.DataFrame(new_data_pd_3)
final_data_1.columns
```

```python
new_data_pd_3.to_json("resume_data.json")
```
